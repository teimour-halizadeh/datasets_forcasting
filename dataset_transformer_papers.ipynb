```python
import pandas as pd

import matplotlib.pyplot as plt

import os
```

```python
root_path = './datasets_former_papers'

def reading_df(datapath):
    path = os.path.join(root_path, datapath)
    print(path)
    df = pd.read_csv(path)


    # Display dataset title
    print(f"\n{'='*50}\nDataset: {datapath}\n{'='*50}")

   # Dataset itself
    display(df.head(n=2))


    

    # print("\nData Types:")
    # display(pd.DataFrame(df.dtypes).T)

    # Display basic information
    print(f"Shape: {df.shape}")
    
    # Display missing values count
    print(f"\nTotal Missing Values: {df.isna().sum().sum()}")

    date = pd.to_datetime(df['date'])

    # Calculate the difference between consecutive dates
    diffs = date.diff().dropna()

    # Get the most common frequency
    most_common_freq = diffs.mode()[0]

    # Print the frequency
    print(f"Frequency of the data is: {most_common_freq}")

    print(f"{'+'*100}")

    return df


    
```

```python

data_list = ["weather.csv", 'electricity.csv', 'traffic.csv', 'national_illness.csv',
             'exchange_rate.csv', 'ETTh1.csv', 'ETTh2.csv', 'ETTm1.csv', 'ETTm2.csv' ]

for data in data_list:
    df = reading_df(data)

    


    
```
