```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
from pathlib import Path
from tqdm import tqdm
```

```python
# We are using already clean data.!
root_path = Path("./pecan_street_data")

path = root_path / '15minute_data_newyork.csv'

df = pd.read_csv(path)
df.head()
```

```python
# We assume that grid is the power they get from the grid
col = ['dataid', 'local_15min', 'grid']
df_newyork = df[col]
df_newyork
```

```python
# the data is clean
df_newyork.isna().sum() 
```

```python
df_newyork.loc[:,'local_15min'] = pd.to_datetime(df_newyork['local_15min'], yearfirst=True)
```

```python
df_prop = df_newyork.groupby('dataid')['local_15min'].agg(['min', 'max', 'count' ])

# check the reqularity of the data
df_prop ['regular_sample_count'] = (df_prop['max'] - df_prop['min'])
td = pd.Timedelta('15min')
df_prop['regular_sample_count'] = df_prop['regular_sample_count']/td + 1
df_prop['length_diff'] = df_prop['count'] - df_prop['regular_sample_count']
display(df_prop)

print('The data is clean, but dataid27 misses a value!')
```

```python

df_pivoted = df_newyork.pivot(values='grid', columns='dataid', index='local_15min')
df_pivoted


```

```python
df_pivoted.isna().sum()
```

```python
# # # fill the only missing value with backward fill
df_pivoted = df_pivoted.fillna(method='backfill')


```
